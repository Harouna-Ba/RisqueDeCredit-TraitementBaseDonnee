```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
```

```python
final_df.head(5)
```

```python
y
```

```python
x
```

```python
X_train,X_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
cf=DecisionTreeClassifier()
cf.fit(X_train, y_train)
```

```python
plot_tree(cf, feature_names=None, class_names=['BON','MAUVAIS'],filled=True)
plt.show()
```

```python
cf.score(X_train,y_train)
```

```python
Y_pred = cf.predict(X_test)
cf.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix = confusion_matrix(y_test, Y_pred)
```

```python
print(confusion_matrix)  
```

#D'apres la matrice de confuision, nous avons 597 + 76 soit 673 prediction correctes et 0 predictions incorrectes

```python
from sklearn.metrics import classification_report 
print(classification_report(y_test, Y_pred))
```

```python
from sklearn.metrics import roc_auc_score 
from sklearn.metrics import roc_curve

 
logit_roc_auc = roc_auc_score(y_test, cf.predict(X_test))

 
fpr, tpr, thresholds = roc_curve(y_test, cf.predict_proba(X_test)[:,1])

 
plt.figure()

 
plt.plot(fpr, tpr, label= '(area = %0.2f)' % logit_roc_auc)

 
plt.plot([0, 1], [0, 1],'r')

 
plt.xlim([0.0, 1.0])

 
plt.ylim([0.0, 1.05])

 
plt.xlabel('False Positive Rate')

 
plt.ylabel('True Positive Rate')

 
plt.title('Receiver operating characteristic')

 
plt.legend(loc='lower right')

 
plt.savefig('Log_ROC')

 
plt.show()
```

```python
%%time
lr = DecisionTreeClassifier()
lr.fit(X_train, y_train)
roc_auc = roc_auc_score(y_test, lr.predict_proba(X_test)[:,1])
print('AUC : %0.4f' %roc_auc)
print(classification_report(y_test, lr.predict(X_test)))
```

D'apres les resultats obtenus, nous avons un AUC = 1. Nous pouvons dire que nous avons un model tres bon, 
Il ne se trompe quasiment pas lors d'une **prediction**
